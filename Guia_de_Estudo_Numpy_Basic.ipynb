## **Importando as Bibliotecas**

```python
# Importe as bibliotecas necessárias # Posteriormente poderemos realizar algumas visualizaçõs gráficas
import numpy as np

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

## 1. Criação de Arrays:

**np.array():** Cria um array a partir de uma lista.

```python
arr = np.array([1,2,3,4,5])

# Imprime o array
print(arr)
```

**np.zeros(), np.ones(), np.full():** Cria arrays de zeros, uns ou com um valor específico.

```python
# Cria um array bidimensional 2x3 preenchido com zeros: 2 linhas x 3 colunas
zeros_arr = np.zeros((2, 3))

# Imprime o array
print(zeros_arr)
```

```python
# Cria um array bidimensional 2x2 preenchido com um: 2 linhas x 2 colunas
ones_arr = np.ones((2, 2))

# Imprime o array
print(ones_arr)
```

```python
# Cria um array unidimensional 2x2 preenchido com um: 2 linhas x 2 colunas
# Uma tupla (3,) array de 3 elementos e depois qual valor será preenchido
full_arr = np.full((3,), 7)
print(full_arr)
```

**np.arange(), np.linspace():** Cria arrays com uma sequência de valores.

```python
# Array com valores de 0 a 10 com passo 2
range_arr = np.arange(0, 10, 2)

# Imprime o array
print(range_arr)
```

```python
# Array de 5 valores espaçados uniformemente entre 0 e 1
linspace_arr = np.linspace(0, 1, 5)

# Imprime o array
print(linspace_arr)
```

## 2. Atributos dos Arrays:

**shape:** Retorna as dimensões do array.

```python
arr = np.array([[1,2,3],[4,5,6]])

# Imprime o array
print(arr.shape)
```

**dtype:** Retorna o tipo de dados no array.

```python
arr = np.array([1.0, 2.5, 3.7])

# Imprime o array
print(arr.dtype)
```

## 3. Indexação e Fatiamento:

Indexação simples e múltipla.

```python
arr = np.array([1,2,3,4,5])

# Imprime o array selecionado
print(arr[2])
```

```python
arr = np.array([1,2,3,4,5])

# Imprime o array fatiado
print(arr[1:4])
```

Indexação em arrays bidimensionais.

```python
arr_2d = np.array([[1,2,3],[4,5,6]])

print(arr_2d[1, 2])
```

## Projeto em Numpy e visualização em Gráficos

Vamos criar um pequeno programa que simula o lançamento de dois dados, calcula a soma dos valores e armazena os resultados em um array. Este é um exemplo prático para entender como o NumPy pode ser usado para lidar com dados numéricos.

```python
def simular_lancamento_dados(n_lancamentos):
    # Simula o lançamento de dois dados 'n_lancamentos' vezes
    resultados = np.random.randint(1, 7, size=(n_lancamentos, 2))
    
    # Calcula a soma dos valores de cada lançamento
    soma_dos_valores = np.sum(resultados, axis=1)
    
    return resultados, soma_dos_valores

# Configurações
num_lancamentos = 10

# Simula o lançamento dos dados
resultados_lancamentos, soma_valores = simular_lancamento_dados(num_lancamentos)

# Mostra os resultados
print("Resultados dos lançamentos:")
print(resultados_lancamentos)

print("\nSoma dos valores em cada lançamento:")
print(soma_valores)
```

```python
def simular_lancamento_dados(n_lancamentos):
    # Simula o lançamento de dois dados 'n_lancamentos' vezes
    resultados = np.random.randint(1, 7, size=(n_lancamentos, 2))
    
    # Calcula a soma dos valores de cada lançamento
    soma_dos_valores = np.sum(resultados, axis=1)
    
    return resultados, soma_dos_valores

# Configurações
num_lancamentos = 100

# Simula o lançamento dos dados
resultados_lancamentos, soma_valores = simular_lancamento_dados(num_lancamentos)

# Gráfico de Barras para a Soma dos Valores
plt.figure(figsize=(10, 5))
plt.hist(soma_valores, bins=np.arange(1.5, 13.5, 1), edgecolor='black')
plt.title('Distribuição da Soma dos Valores')
plt.xlabel('Soma dos Valores')
plt.ylabel('Frequência')
plt.show()

# Boxplot para a Soma dos Valores
plt.figure(figsize=(8, 5))
sns.boxplot(x=soma_valores, color='lightblue')
plt.title('Boxplot da Soma dos Valores')
plt.xlabel('Soma dos Valores')
plt.show()

# Gráfico de Dispersão para os Resultados dos Lançamentos
plt.figure(figsize=(8, 6))
plt.scatter(resultados_lancamentos[:, 0], resultados_lancamentos[:, 1], color='orange')
plt.title('Lançamento de Dois Dados')
plt.xlabel('Dado 1')
plt.ylabel('Dado 2')
plt.show()

# Pair Plot para visualizar relações entre variáveis
sns.pairplot(pd.DataFrame(resultados_lancamentos, columns=['Dado 1', 'Dado 2']))
plt.suptitle('Pair Plot dos Resultados dos Lançamentos', y=1.02)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

def simular_carteira(retornos_diarios, capital_inicial=100000):
    num_ativos = retornos_diarios.shape[1]
    num_dias = retornos_diarios.shape[0]

    # Gera pesos aleatórios para os ativos
    pesos = np.random.rand(num_ativos)
    pesos /= np.sum(pesos)  # Normaliza para garantir que a soma seja 1

    # Inicializa o valor da carteira ao longo do tempo
    valor_carteira = np.zeros(num_dias)
    valor_carteira[0] = capital_inicial

    for dia in range(1, num_dias):
        # Calcula o retorno diário da carteira
        retorno_diario = np.sum(pesos * retornos_diarios[dia])

        # Atualiza o valor da carteira
        valor_carteira[dia] = valor_carteira[dia - 1] * (1 + retorno_diario)

    return valor_carteira

# Configurações
num_dias_simulacao = 252  # Número de dias de negociação em um ano
num_ativos = 5  # Número de ativos na carteira

# Gera retornos diários aleatórios para cada ativo
retornos_diarios = np.random.normal(0.0005, 0.02, size=(num_dias_simulacao, num_ativos))

# Simula a carteira
valor_carteira_simulada = simular_carteira(retornos_diarios)

# Visualiza o desempenho da carteira
plt.figure(figsize=(10, 6))
plt.plot(valor_carteira_simulada, label='Carteira Simulada')
plt.title('Desempenho da Carteira de Investimentos')
plt.xlabel('Dias de Negociação')
plt.ylabel('Valor da Carteira')
plt.legend()
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import yfinance as yf
import datetime

import plotly.graph_objects as go
from plotly.subplots import make_subplots

def obter_dados(tickers, inicio, fim):
    # Obtendo dados do Yahoo Finance
    dados = yf.download(tickers, start=inicio, end=fim)['Adj Close']
    return dados

def simular_carteira(dados, capital_inicial=100000):
    num_ativos = dados.shape[1]
    num_dias = dados.shape[0]

    # Gera pesos aleatórios para os ativos
    pesos = np.random.rand(num_ativos)
    pesos /= np.sum(pesos)  # Normaliza para garantir que a soma seja 1

    # Inicializa o valor da carteira ao longo do tempo
    valores_carteira = pd.DataFrame(index=dados.index, columns=['Valor_Carteira'])
    valores_carteira['Valor_Carteira'].iloc[0] = capital_inicial

    for dia in range(1, num_dias):
        # Calcula o retorno diário da carteira
        retorno_diario = np.sum(pesos * (dados.iloc[dia] / dados.iloc[dia - 1] - 1))

        # Atualiza o valor da carteira
        valores_carteira['Valor_Carteira'].iloc[dia] = valores_carteira['Valor_Carteira'].iloc[dia - 1] * (1 + retorno_diario)

    return valores_carteira, pesos

# Configurações
tickers = ['AAPL', 'GOOGL', 'MSFT', 'AMZN', 'TSLA']  # Símbolos de ações
inicio_data = datetime.datetime(2023, 1, 1)
fim_data = datetime.datetime(2023, 12, 31)

# Obtém dados para o ano de 2023
dados_2023 = obter_dados(tickers, inicio_data, fim_data)

# Simula a carteira
capital_inicial = 100000  # Defina o capital inicial aqui
valores_carteira_simulados, pesos = simular_carteira(dados_2023, capital_inicial)

# Visualiza o desempenho da carteira
plt.figure(figsize=(10, 6))
plt.plot(valores_carteira_simulados, label='Carteira Simulada')
plt.title('Desempenho da Carteira de Investimentos (2023)')
plt.xlabel('Data')
plt.ylabel('Valor da Carteira')
plt.legend()

# Ajusta o formato do eixo x para mostrar apenas as datas
plt.gca().xaxis.set_major_formatter(plt.matplotlib.dates.DateFormatter('%Y-%m-%d'))

# Incluindo a fonte dos dados
plt.figtext(0.99, 0.01, 'Fonte dos Dados: Yahoo Finance', horizontalalignment='right', fontsize=8, color='gray')

plt.show()

# Adicione ao código existente
pesos_ativos = (dados_2023 / dados_2023.iloc[0]) * pesos

plt.figure(figsize=(10, 6))
plt.stackplot(pesos_ativos.index, pesos_ativos.values.T, labels=dados_2023.columns)
plt.title('Alocação de Ativos na Carteira de Investimentos (2023)')
plt.xlabel('Data')
plt.ylabel('Alocação de Peso')
plt.legend(loc='upper left', bbox_to_anchor=(1, 1))
plt.gca().xaxis.set_major_formatter(plt.matplotlib.dates.DateFormatter('%Y-%m-%d'))
plt.show()

# Adicione ao código existente
retorno_cumulativo = (valores_carteira_simulados / capital_inicial - 1) * 100

plt.figure(figsize=(10, 6))
plt.plot(retorno_cumulativo, label='Retorno Cumulativo')
plt.title('Retorno Cumulativo da Carteira de Investimentos (2023)')
plt.xlabel('Data')
plt.ylabel('Retorno Cumulativo (%)')
plt.legend()
plt.gca().xaxis.set_major_formatter(plt.matplotlib.dates.DateFormatter('%Y-%m-%d'))

```

```python
import plotly.graph_objects as go
import pandas as pd
import yfinance as yf
from datetime import datetime, timedelta
from plotly.subplots import make_subplots

# Função para obter dados em tempo real
def obter_dados_tempo_real(tickers):
    dados = yf.download(tickers, start=datetime.now() - timedelta(days=30), end=datetime.now())['Adj Close']
    return dados

# Função para atualizar gráfico em tempo real
def atualizar_grafico(fig, dados_tempo_real):
    for coluna in dados_tempo_real.columns:
        fig.add_trace(go.Scatter(x=dados_tempo_real.index, y=dados_tempo_real[coluna], mode='lines', name=coluna))
    fig.update_layout(title='Desempenho em Tempo Real',
                      xaxis_title='Data',
                      yaxis_title='Preço de Fechamento',
                      showlegend=True)
    return fig

# Configurações
tickers = ['AAPL', 'GOOGL', 'MSFT', 'AMZN', 'TSLA']

# Obtém dados iniciais
dados_iniciais = obter_dados_tempo_real(tickers)

# Cria o layout do gráfico
fig = make_subplots(rows=1, cols=1)

# Atualiza o gráfico em tempo real
fig = atualizar_grafico(fig, dados_iniciais)

# Mostra o gráfico interativo
fig.show()
```
